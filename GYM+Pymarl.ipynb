```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load your dataset
df = pd.read_csv("/content/MARL_6G_results.csv")

# Clean column names (remove whitespaces if any)
df.columns = df.columns.str.strip()

# ✔ Rename for ease if needed (optional)
# df.rename(columns={'energy_efficiency_bits_per_joule': 'energy_efficiency'}, inplace=True)

# 1. 3D Plot: timestep vs avg_reward vs avg_throughput_bps
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_trisurf(df['timestep'], df['avg_reward'], df['avg_throughput_bps'], cmap='viridis')
ax1.set_xlabel('Timestep')
ax1.set_ylabel('Avg Reward')
ax1.set_zlabel('Throughput (bps)')
plt.title('3D: timestep vs avg_reward vs avg_throughput_bps')
plt.tight_layout()
plt.show()

# 2. 3D Plot: timestep vs avg_latency_ms vs avg_throughput_bps
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.plot_trisurf(df['timestep'], df['avg_latency_ms'], df['avg_throughput_bps'], cmap='plasma')
ax2.set_xlabel('Timestep')
ax2.set_ylabel('Latency (ms)')
ax2.set_zlabel('Throughput (bps)')
plt.title('3D: timestep vs avg_latency_ms vs avg_throughput_bps')
plt.tight_layout()
plt.show()

# 3. 3D Plot: timestep vs avg_fairness vs avg_interference
fig3 = plt.figure()
ax3 = fig3.add_subplot(111, projection='3d')
ax3.plot_trisurf(df['timestep'], df['avg_fairness'], df['avg_interference'], cmap='coolwarm')
ax3.set_xlabel('Timestep')
ax3.set_ylabel('Fairness')
ax3.set_zlabel('Interference')
plt.title('3D: timestep vs avg_fairness vs avg_interference')
plt.tight_layout()
plt.show()

# 4. 3D Plot: timestep vs energy_efficiency_bits_per_joule vs avg_throughput_bps
fig4 = plt.figure()
ax4 = fig4.add_subplot(111, projection='3d')
ax4.plot_trisurf(df['timestep'], df['energy_efficiency_bits_per_joule'], df['avg_throughput_bps'], cmap='cividis')
ax4.set_xlabel('Timestep')
ax4.set_ylabel('Energy Efficiency')
ax4.set_zlabel('Throughput (bps)')
plt.title('3D: timestep vs energy_efficiency vs throughput')
plt.tight_layout()
plt.show()

# 5. 3D Plot: timestep vs avg_reward vs energy_efficiency_bits_per_joule
fig5 = plt.figure()
ax5 = fig5.add_subplot(111, projection='3d')
ax5.plot_trisurf(df['timestep'], df['avg_reward'], df['energy_efficiency_bits_per_joule'], cmap='inferno')
ax5.set_xlabel('Timestep')
ax5.set_ylabel('Avg Reward')
ax5.set_zlabel('Energy Efficiency')
plt.title('3D: timestep vs avg_reward vs energy_efficiency')
plt.tight_layout()
plt.show()
```
